# Import required Libraries

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80,random_state=3)
```

```python
plt.scatter(X,y)
```

```python
reg=LinearRegression()
reg.fit(X,y)
print("m: ",reg.coef_)
print("b: ",reg.intercept_)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

# 1st iteration or Epoch

```python
# Lets apply Gradient Decent assuming slope is constant m= 93.55
# and let's assume the startign value for intercept b = 200
y_pred = ((93.55*X)+200).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='black',label='b=200')
plt.legend()
plt.show()
```

```python
m=93.55
b=200

loss_slope = -2*np.sum(y-m*X.ravel() - b)
loss_slope
```

```python
# Lets take Learning Rate = 0.1
lr = 0.1

step_size = loss_slope * lr
step_size
```

```python
# Calculating the new intercept
b = b - step_size
b
```

```python
y_pred1= ((93.55 * X) + b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='black',label='b= 54.94')
plt.plot(X,y_pred,color='#808080',label='b= 200')
plt.legend()
plt.show
```

# 2nd Iteration or Epoch

```python
m= 93.55
b= 54.94

loss_slope = -2*np.sum(y-m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope * lr
step_size
```

```python
b = b-step_size
b
```

```python
y_pred2 = ((93.55 * X) + b).reshape(4)
plt.figure(figsize=(12,10))
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='black',label='b=25.93')
plt.plot(X,y_pred1,color='#808080',label='b=54.94')
plt.plot(X,y_pred,color='#d9d9d9',label='b=200')
plt.legend()
plt.show()
```

# 3rd Iteration or Epoch

```python
loss_slope= -2*np.sum(y-m*X.ravel() -b)
loss_slope
```

```python
step_size= loss_slope*lr
step_size
```

```python
b= b-step_size
b
```

```python
y_pred3 = ((93.55 * X) + b).reshape(4)
```

```python
plt.figure(figsize=(12,10))
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred3,color='black',label='20.12')
plt.plot(X,y_pred2,color='#666666',label='b=25.93')
plt.plot(X,y_pred1,color='#999999',label='b=54.94')
plt.plot(X,y_pred,color='#cccccc',label='b=200')
plt.legend()
plt.show()
```

# for loop is used to perform 10 iterations or epochs

```python
plt.figure(figsize=(12,10))
b= 200
m= 93.55
lr= 0.1

epochs=10

for i in range(epochs):
  loss_slope=-2*np.sum(y-m*X.ravel() -b)
  b= b-(lr* loss_slope)
  y_pred= m*X+b
  plt.plot(X,y_pred)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='black')
plt.show()
```

# If you choose a learning rate that is too small, it will take a longer time to reach the OLS, depending on the number of epochs you select

```python
plt.figure(figsize=(12,10))
b= 200
m= 93.55
lr= 0.01

epochs=20

for i in range(epochs):
  loss_slope=-2*np.sum(y-m*X.ravel() -b)
  b= b-(lr* loss_slope)
  y_pred= m*X+b
  plt.plot(X,y_pred)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='black')
plt.show()
```
